# Nataf Transformation: Example 3 - Correlation Distortion
Calculate correlation distortion for the transformation of two random variables from normal to lognormal.

Author: Michael D. Shields

Date: 7/15/2018

```python
from UQpy.SampleMethods import Nataf
import numpy as np
import matplotlib.pyplot as plt
```

Compute the correlation distortion at various values of Gaussian correlation

```python
w3 = np.zeros(101)
rho = np.linspace(-0.999, 0.999, 101)
for i in range(101):
    Rho1 = np.array([[1.0, rho[i]], [rho[i], 1.0]])
    ww = Nataf(corr_norm=Rho1, dist_name=['Lognormal', 'Lognormal'], dist_params=[[0, 1], [0, 1]], dimension=2)
    w3[i] = ww.corr[0, 1]
    print(i)
```

Plot the correlation distortion along with the straight-line associated with no distortion

```python
plt.plot(rho, w3)
plt.plot(rho, rho)
plt.show()
```
