In this example, Stratified sampling is used to generate samples from Uniform probability distribution and sample are added using adaptive approach Refined Stratified Sampling.

Import the necessary libraries. Here we import standard libraries such as numpy and matplotlib, but also need to import the STS and RSS class from UQpy.

```python
from UQpy.SampleMethods import STS
from UQpy.SampleMethods import RSS
from UQpy.Distributions import Uniform
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
from scipy.spatial import Delaunay
```

Create a distribution object.

```python
marginals = [Uniform(loc=0., scale=1.), Uniform(loc=0., scale=1.)]
```

Run stratified sampling.

```python
x = STS(dist_object=marginals, stype='Voronoi', nsamples=16)
```

Using UQpy RSS class to expand samples generated by STS class. In this example, two new samples are generated inside cells with maximum weight associated with it.

```python
y = RSS(sample_object=x, nsamples=18, n_add=2)
```

In the first figure shown below, samples generated from STS class are plotted. Second figure, shows the voronoi tesselation with initial samples (i.e. 16 samples). Third figure, shows the voronoi tesselation with all the samples. It can be observe that new sample is randomly added inside the voronoi cell with maximum area.

```python
fig1 = plt.figure()
ax = fig1.gca()
ax.set_xticks(np.arange(0, 1, 0.25))
ax.set_yticks(np.arange(0, 1, 0.25))
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.plot(y.sample_object.samples[:16, 0], y.sample_object.samples[:16, 1], 'ro')
boundary_vertices = y.mesh_vertices[y.points_to_samplesU01 == -1, :]
samples16 = y.mesh_vertices[boundary_vertices.shape[0]:boundary_vertices.shape[0]+16, :]
points = np.vstack([samples16, boundary_vertices])
tri = Delaunay(points)
plt.triplot(points[:, 0], points[:, 1], tri.simplices)
plt.grid()
plt.show()

fig2 = plt.figure()
ax = fig2.gca()
ax.set_xticks(np.arange(0, 1, 0.25))
ax.set_yticks(np.arange(0, 1, 0.25))
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.plot(y.sample_object.samples[:16, 0], y.sample_object.samples[:16, 1], 'ro')
samples18 = y.mesh_vertices[boundary_vertices.shape[0]:boundary_vertices.shape[0]+18, :]
points = np.vstack([samples18, boundary_vertices])
tri1 = Delaunay(points)
plt.triplot(points[:, 0], points[:, 1], tri1.simplices)
plt.plot(y.sample_object.samples[16:18, 0], y.sample_object.samples[16:18, 1], 'go')
plt.grid()
plt.show()
```

Further, RSS class is used to adaptively increase the sample size. In this example, samples are randomly added in cell with maximum weights associated with it and new sample generated using Simplex class.

```python
y.run(nsamples=50)
```

In the figure shown below, all samples generated from STS and RSS class are plotted.

```python
fig3 = plt.figure()
ax = fig3.gca()
ax.set_xticks(np.arange(0, 1, 0.25))
ax.set_yticks(np.arange(0, 1, 0.25))
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.plot(y.sample_object.samples[:, 0], y.sample_object.samples[:, 1], 'ro')
tri2 = Delaunay(y.mesh_vertices)
plt.triplot(y.mesh_vertices[:, 0], y.mesh_vertices[:, 1], tri2.simplices)
plt.plot(y.sample_object.samples[16:, 0], y.sample_object.samples[16:, 1], 'go')
plt.grid()
plt.show()
```
